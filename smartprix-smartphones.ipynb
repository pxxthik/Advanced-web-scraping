```python
with open("smartprix.html", "r", encoding="utf-8") as f:
    html = f.read()
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
soup = BeautifulSoup(html, "lxml")
```

```python
containers = soup.find_all("div", {"class": "sm-product has-tag has-features has-actions"})
```

```python
name = []
price = []
spec_score = []
sim = []
processor = []
ram = []
battery = []
display = []
camera = []
card = []
os = []

for i in soup.find_all("div", {"class": "sm-product has-tag has-features has-actions"}):
    
    try:
        name.append(i.find('h2').text)
    except:
        name.append(np.nan)
    
    try:
        price.append(i.find('span', {'class': 'price'}).text)
    except:
        price.append(np.nan)
        
    try:
        spec_score.append(i.find('div', {'class': 'score rank-2-bg'}).find('b').text)
    except:
        spec_score.append(np.nan)
    
    x = i.find('ul', {"class": "sm-feat specs"}).find_all('li')
    
    try:
        sim.append(x[0])
    except:
        sim.append(np.nan)
    
    try:
        processor.append(x[1])
    except:
        processor.append(np.nan)
    
    try:
        ram.append(x[2])
    except:
        ram.append(np.nan)
        
    try:
        battery.append(x[3])
    except:
        battery.append(np.nan)
    
    try:
        display.append(x[4])
    except:
        display.append(np.nan)
    
    try:
        camera.append(x[5])
    except:
        camera.append(np.nan)
    
    try:
        card.append(x[6])
    except:
        card.append(np.nan)
    
    try:
        os.append(x[7])
    except:
        os.append(np.nan)
```

```python
df = pd.DataFrame({
    'model': name,
    'price': price,
    'score': spec_score,
    'sim': sim,
    'processor': processor,
    'ram': ram,
    'battery': battery,
    'display': display,
    'camera': camera,
    'card': card,
    'os': os
})

df
```
